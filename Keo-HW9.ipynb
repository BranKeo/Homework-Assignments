```python
#Let's Make A Deal

#randomly choose to switch/keep initial door
import numpy as np
Num_wins = 0
Num_wins_switch = 0
Num_wins_keep = 0
Num_losses = 0
Num_losses_switch = 0
Num_losses_keep = 0

#randomly put prize in door
for game in range(0, 1001):
    if game < 1000:
        prize = np.random.random(1)

        if prize < .333333333:
            prize_doorA = True
            prize_doorB = False
            prize_doorC = False
        elif prize < .666666666 and prize > .333333333:
            prize_doorB = True
            prize_doorA = False
            prize_doorC = False
        else:
            prize_doorC = True
            prize_doorA = False
            prize_doorB = False

        #randomly choose door
        choose_door = np.random.random(1)

        if choose_door < .333333333:
            choose_doorA = True
            choose_doorB = False
            choose_doorC = False
        elif choose_door < .666666666 and choose_door > .333333333:
            choose_doorB = True
            choose_doorA = False
            choose_doorC = False
        else:
            choose_doorC = True
            choose_doorA = False
            choose_doorB = False

        #eliminate one of 2 unpicked doors
        #if both doors no prize
        random_elim = np.random.random(1)
        if prize_doorA is True and choose_doorA is True:
            if random_elim < .5:
                elim_doorB = True
                elim_doorC = False
            else:
                elim_doorB = False
                elim_doorC = True
        elif prize_doorB is True and choose_doorA is True:
            if random_elim < .5:
                elim_doorA = True
                elim_doorC = False
            else:
                elim_doorA = False
                elim_doorC = True
        elif prize_doorC is True and choose_doorA is True:
            if random_elim < .5:
                elim_doorA = True
                elim_doorB = False
            else:
                elim_doorA = False
                elim_doorB = True
        #if one door has prize
        elif prize_doorA is True and choose_doorB is True:
            elim_doorC = True
        elif prize_doorA is True and choose_doorC is True:
            elim_doorB = True
        elif prize_doorB is True and choose_doorA is True:
            elim_doorC = True
        elif prize_doorB is True and choose_doorC is True:
            elim_doorA = True
        elif prize_doorC is True and choose_doorA is True:
            elim_doorB = True
        elif prize_doorC is True and choose_doorB is True:
            elim_doorA = True

        #randomly choose to switch/keep initial door
        switch1 = np.random.random(1)
        if elim_doorA is True and choose_doorB is True:
            if switch1 > .5:
                switched = False
            else:
                choose_doorB = False
                choose_doorC = True
                switched = True
        elif elim_doorA is True and choose_doorC is True:
            if switch1 > .5:
                switched = False
            else:
                choose_doorC = False
                choose_doorB = True
                switched = True
        elif elim_doorB is True and choose_doorC is True:
            if switch1 > .5:
                switched = False
            else:
                choose_doorC = False
                choose_doorA = True
                switched = True
        elif elim_doorB is True and choose_doorA is True:
            if switch1 > .5:
                switched = False
            else:
                choose_doorA = False
                choose_doorC = True
                switched = True
        elif elim_doorC is True and choose_doorA is True:
            if switch1 > .5:
                switched = False
            else:
                choose_doorA = False
                choose_doorB = True
                switched = True
        elif elim_doorC is True and choose_doorB is True:
            if switch1 > .5:
                switched = False
            else:
                choose_doorB = False
                choose_doorA = True
                switched = True

        
        #Decide if you won and if win was from switching
        if choose_doorA is True and prize_doorA is True:
            winner = True
            Num_wins += 1
            if switched is True:
                Num_wins_switch += 1
            else:
                Num_wins_keep += 1
        elif choose_doorB is True and prize_doorB is True:
            winner = True
            Num_wins += 1
            if switched is True:
                Num_wins_switch += 1
            else:
                Num_wins_keep += 1
        elif choose_doorC is True and prize_doorC is True:
            winner = True
            Num_wins += 1
            if switched is True:
                Num_wins_switch += 1
            else:
                Num_wins_keep += 1
        else:
            winner = False
            Num_losses += 1
            if switched is True:
                Num_losses_switch += 1
            else:
                Num_losses_keep += 1
        

    else:
        print('Total number of wins:',str(Num_wins))
        print('Number of wins by switching:',str(Num_wins_switch))
        print('Number of wins by keeping:',str(Num_wins_keep))
        print('')
        print('Total number of losses:',str(Num_losses))
        print('Number of losses by switching:',str(Num_losses_switch))
        print('Number of losses by keeping:',str(Num_losses_keep))
        print('')



print('Probability of winning by switching:',str(Num_wins_switch/1000))
print('Probability of winning by keeping:',str(Num_wins_keep/1000))
print('Probability of winning appears to be about the same for switching and keeping')
```
